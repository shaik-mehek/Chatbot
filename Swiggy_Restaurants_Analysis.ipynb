## Data Cleaning and Preprocessing:

#### 1. Remove Duplicates:

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

#### 2. Handle Missing Values:

```python
# Fill missing Categorical/Text columns
df['Cuisine'] = df['Cuisine'].fillna('Unknown')
```

```python
df['Offer Name'] = df['Offer Name'].fillna('No Offer')
```

```python
df['Area'] = df['Area'].fillna('Unknown Area')
```

```python
# Fill missing Numeric column
df['Number of Ratings'] = df['Number of Ratings'].fillna(0)
```

```python
# fill NaN values with 0 in Rating column
df['Rating'] = df['Rating'].fillna(0)
```

#### 3. Text Cleaning:

```python
#The restaurants with rating '--' is replaced with zero
df['Rating'] = df['Rating'].replace('--', '0')
```

```python
df['Number of Ratings'] = df['Number of Ratings'].replace('Too Few Ratings', '0')
```

```python
df['Offer Name'] = df['Offer Name'].str.replace('\n', ',')
```

#### 4. Rename column:

```python
# Rename a single column
df.rename(columns={'Average Price': 'Average Price(Rs)'}, inplace=True)
```

#### 5. Convert Data Type:

```python
# Convert 'Rating' to float
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Remove non-numeric characters and convert to numeric
df['Average Price(Rs)'] = pd.to_numeric(df['Average Price(Rs)'].astype(str).str.replace('[^\d]', '', regex=True), errors='coerce')
```

#### 6. String Cleaning in Columns:

```python
# Remove ₹ and " for two" from 'Average Price'
df['Average Price(Rs)'] = df['Average Price(Rs)'].astype(str).str.replace('₹', '', regex=False)
df['Average Price(Rs)'] = df['Average Price(Rs)'].astype(str).str.replace(' for two', '', regex=False)
```

```python
# Convert columns to title case
df['Offer Name'] = df['Offer Name'].str.title()
df['Restaurant Name']=df['Restaurant Name'].str.title()
df['Area']=df['Area'].str.title()
```

#### Pure Veg vs Non-Veg Restaurants

#### Pure Veg and Non-Veg Restaurant in Top 5 Cities

```python
# Filter for pure veg restaurants
pure_veg_df = df[df['Pure Veg'] == 'Yes']

# Filter for non-veg restaurants
non_veg_df = df[df['Pure Veg'] == 'No']

# Count restaurants by location for pure veg
pure_veg_counts = pure_veg_df['Location'].value_counts().reset_index()
pure_veg_counts.columns = ['Location', 'Pure Veg Count']

# Count restaurants by location for non-veg
non_veg_counts = non_veg_df['Location'].value_counts().reset_index()
non_veg_counts.columns = ['Location', 'Non-Veg Count']

# Merge the counts
merged_counts = pd.merge(pure_veg_counts, non_veg_counts, on='Location', how='outer').fillna(0)
```

```python
# Pure Veg and Non-Veg Restaurant in Top 5 Cities
merged_counts['Total Count'] = merged_counts['Pure Veg Count'] + merged_counts['Non-Veg Count']
top_5_cities_total = merged_counts.sort_values(by='Total Count', ascending=False).head(5)

# Reshape the data for grouped bar plot
melted_counts = top_5_cities_total.melt(id_vars='Location', value_vars=['Pure Veg Count', 'Non-Veg Count'],
                                        var_name='Veg Type', value_name='Count')

plt.figure(figsize=(12, 6))
sns.barplot(x='Location', y='Count', hue='Veg Type', data=melted_counts, palette={'Pure Veg Count': 'lightgreen', 'Non-Veg Count': 'salmon'})
plt.title('Pure Veg and Non-Veg Restaurant in Top 5 Cities', fontsize=14, fontweight='bold')
plt.xlabel('Location', fontsize=12, fontweight='bold')
plt.ylabel('Number of Restaurants', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')

# Create custom legend handles with color representation
from matplotlib.patches import Patch
legend_handles = [Patch(facecolor='lightgreen', label='Pure Veg'),
                  Patch(facecolor='salmon', label='Non-Veg')]

plt.legend(handles=legend_handles, title='Vegetarian Type', loc='upper left')
plt.tight_layout()
plt.show()
```

#### Rating Distribution for Top 5 Locations

```python
# Get top 5 locations with most restaurants
top5_locations = df['Location'].value_counts().nlargest(5).index

# Filter the dataframe
top5_df = df[df['Location'].isin(top5_locations)]

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=top5_df, x='Location', y='Rating', palette='Set2')
plt.title('Violin Plot: Rating Distribution for Top 5 Locations')
plt.xlabel('Location')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()

```

#### Average Price Trend by Location (Top 5 Locations)

```python

# Get top 5 restaurants by Average Price(Rs)
top5_restaurants = df[['Restaurant Name', 'Average Price(Rs)']].groupby('Restaurant Name').mean().nlargest(5, 'Average Price(Rs)')

# Sort for better visualization
top5_restaurants = top5_restaurants.sort_values(by='Average Price(Rs)')

# Plot
plt.figure(figsize=(10, 6))
top5_restaurants.plot(kind='area', color='lightgreen', alpha=0.7, legend=False)
plt.title('Average Price of Top 5 Restaurants (Area Plot)')
plt.ylabel('Average Price')
plt.xlabel('Restaurant Name')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


```
